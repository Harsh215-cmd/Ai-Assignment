# Strings Methods

```python
#Upper

str1 ="Harshal"
str_upper=str1.upper()
print(str_upper)
```

```python
#is alpha

gtreeting="Helloworld";
print(gtreeting.isalpha())

gtreeting2="Helloworld1";
print(gtreeting2.isalpha())
```

```python
#isdigit

num1 = "100";
print(num1.isdigit())

num2 = "hundred";
print(num2.isdigit())
```

```python
#isalnum

str1="Harsh123"
print(str1.isalnum())
```

```python
# islower

str1="harshal"
print(str1.islower())
```

```python
#is Upper

str1="harsh"
print(str1.isupper())

str2 = str1.upper()

print(str2.isupper())
```

```python
# startswith

str1="hello my name is harsh"
print(str1.startswith("hello"))
print(str1.startswith("harsh"))
```

```python
#endwith

str1="hello my name is harsh"
print(str1.endswith("hello"))
print(str1.endswith("harsh"))
```
